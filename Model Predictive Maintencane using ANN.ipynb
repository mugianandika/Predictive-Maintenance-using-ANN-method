# Normalisasi dan Rolling Mean

```python
# Dataframe feats untuk drop kolom RUL, minutes, label_bnc, dan kategori pada data train
feats = train.columns.drop(['RUL','minutes','label_bnc','kategori'])
```

```python
# Size window rolling mean yang dipilih
ws=5

#Rolling mean untuk menghaluskan fluktuasi jangka pendek data
train[feats] = train[feats].rolling(5, min_periods=1).mean()

# Melihat lima baris pertama data train
train.head()
```

```python
# Melakukan normalisasi data train tanpa kolom RUL, minutes, label_bnc, dan kategori
sc = MinMaxScaler(feature_range=(-1,1))
train[feats] = sc.fit_transform(train[feats])
```

```python
# Melihat lima baris pertama data train
train.head()
```

```python
# Plotting Data Train
axes = train.plot.line(subplots=True,figsize=(10,15))
type(axes)
```

```python
# Memisahkan data train menjadi X_train dan y_train
# Menghapus kolom 'label_bnc','RUL','minutes','kategori' dari data train
X_train = train.drop(['label_bnc','RUL','minutes','kategori'],axis=1)
# Menyimpan kolom 'kategori' pada data train kedalam variabel y_train
y_train = train['kategori']
```

# Metode Artificial Neural Network (ANN)

```python
# Memisahkan data dari X_train dan y_train menjadi x_train dan x_valid ; y_train dan y_valid dengan data train=80% dan data validasi=20%
x_train, x_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=0,stratify=y_train)
```

```python
# Menginisialisasikan Neural Network
model = Sequential()

# Membuat layer dengan fungsi aktivasi relu untuk input dan hidden layer dan fungsi aktivasi sigmoid untuk output layer
model.add(Dense(9, kernel_initializer = 'uniform', activation = 'relu',input_dim = 8))
model.add(Dense(9, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(5, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Ringkasan model
model.summary()
```

```python
# Compile Neural Network
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Melatih model menggunakan ANN dengan perulangan 28
training=model.fit(x_train, y_train, batch_size = 32, epochs = 28, validation_data=(x_valid, y_valid))
```

```python
# Evaluasi train score dan train accuracy
score, acc = model.evaluate(x_train, y_train,batch_size=32) 
print('Train score:', score) 
print('Train accuracy:', acc) 
```

```python
# Menyimpan history loss dari data training ke dalam variabel loss
loss = training.history['loss']

# Menyimpan history validasi loss kedalam variabel val_loss
val_loss = training.history['val_loss']

# Plotting loss training dan validasi
epochs = range(1, 29)
plt.plot(epochs, loss, 'bo', label='Training_loss')
plt.plot(epochs, val_loss, 'b', label='Validation_loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Plotting training accuracy dan validation accuracy
plt.clf()
acc = training.history['accuracy']
val_acc = training.history['val_accuracy']
plt.plot(epochs, acc, 'bo', label='Training_accuracy')
plt.plot(epochs, val_acc, 'b', label='Validaion_accuracy')
plt.legend()
plt.show()
```
